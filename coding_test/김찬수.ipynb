1. P511 배낭함수 문제의 1, 2에 적합한 코드를 골라서 작성하시오.

```python
maxWeight = 7 # 최대 무게
rowCount = 4 # 행의 갯수
jem = ['', '금괴', '수정', '루비', '진주'] # 보석 리스트
weight = [0, 6, 4, 3, 5]  # 보석무게[0, 금괴, 수정, 루비, 진주]
money = [0, 13, 8, 6, 12]  # 보석가격[0, 금괴, 수정, 루비, 진주]
# 함수 선언

# 위에는 제가 그냥 보려고 넣었습니다.

def knapsack():
    # 배열을 maxWeight(행) + 1와 rowCount(열) + 1 크기만큼 생성한다
    array = [[0 for _ in range(maxWeight + 1)] for _ in range(rowCount+1)]
    for row in range(1, rowCount + 1):              # 행의 갯수
        for col in range(1, maxWeight + 1):         # 열의 갯수
             # 행이 열보다 크면 바로 위 행의 결과를 가져온다
            if weight[row] > col:
                array[row][col] = array[row-1][col]
            else: # 무게가 가방무게보다 작거나 같으면
                # 현재보석 가치와 배낭무게에서 현재보석 무게를 뺀 위치의 가격을 합산한다
                value1 = money[row] + array[row-1][col-weight[row]] # 현재 보석이 들어간 가치
                value2 = array[row-1][col] # 현재보석을 넣지 않았을때의 가치
                array[row][col] = max(value1, value2) # 둘중에 더 큰것을 선택

    return array[rowCount][maxWeight] # 최종결과값
```

2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오.

```python
# 유명 맛집 대기줄

## 함수선언
# isQueueFull
#----------------------------------------------------------------------------------------
def isQueueFull():
    global SIZE, queue, front, rear
    if rear == SIZE - 1: # 큐가 꽉찼는지 확인
        return True
    else:
        return False
#----------------------------------------------------------------------------------------

# isQueueEmpty
#----------------------------------------------------------------------------------------
def isQueueEmpty():
    global SIZE, queue, front, rear
    if (front == rear): # 큐가 비었는지 확인
        return True
    else:
        return False
#----------------------------------------------------------------------------------------

# enQueue함수
#----------------------------------------------------------------------------------------
def enQueue(data):
    global SIZE, queue, front, rear
    if isQueueFull(): # 큐가 가득 차있다면 가득참을 반환
        print('Queue is full')
        return
    rear += 1 # rear를 1증가시켜 데이터가 들어갈 위치를 지정
    queue[rear] = data # 증가한 rear자리에 데이터 저장
    print(f'{data}을(를) 대기줄에 추가')
#----------------------------------------------------------------------------------------

# deQueue함수
#----------------------------------------------------------------------------------------
def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty(): # 큐가 비어있다면 비었다는것을 반환
        print('Queue is empty')
        return None
    front += 1 # 가장 앞자리를 가리킴
    data = queue[front] # 증가된 front자리에서 데이터를 앞으로 가져옴
    queue[front] = None # 삭제된 자리 초기화
    
    # 모든 사람을 한 칸씩 앞으로 당김
    for i in range(front+1, rear+1):
        queue[i-1] = queue[i]
        queue[i] = None

    front -= 1
    rear -= 1
    return data
#----------------------------------------------------------------------------------------

# peek함수
#----------------------------------------------------------------------------------------
def peek(): # 큐의 앞을 확인
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('Queue is empty')
        return None
    return queue[front+1] # front 바로뒤의 인덱스 
#----------------------------------------------------------------------------------------

# 초기화
#----------------------------------------------------------------------------------------
SIZE = 5
queue = [None for _ in range(SIZE)] # 큐를 SIZE에 맞게 초기화
front = rear = -1
#----------------------------------------------------------------------------------------

## 메인 모듈
#----------------------------------------------------------------------------------------
while True:
    print('\n[메뉴 선택]')
    print('1: 손님 추가')
    print('2: 손님 입장')
    print('3: 종료')
    print(f'현재 대기 줄 상태: {queue}')

    choice = input('메뉴를 선택하세요: ')

    if choice == '1':    # '1'번을 입력하면 손님 추가
        name = input('추가할 이름을 입력하세요: ')
        enQueue(name)
    
    elif choice == '2':  # '2'번을 입력하면 손님 입장(삭제)
        deQueue()

    elif choice == '3':  # '3'번을 입력하면 종료
        print('식당 영업 종료!')
        break

    else: # 예외처리 : 잘못된 것 을 고르면 다시
        print('잘못된 입력입니다. 다시 고르세요.')
#----------------------------------------------------------------------------------------
```

3. P309 편의점 판매된 물건 목록 출력하기 문제 역시 while True: 문으로 동적으로 입력받는 프로그램으로 변경하시오.
데이터의 종류는 [ 콘말차, 삿포로 맥주, 아카페라 벤티 헤이즐넛, 레어치즈푸딩, 척오리지널 메가 사워, 요아정 요거트컵, 페퍼로니피자주먹밥, 널담 슈톨렌, 딸기마시멜로케이크, 버니공쥬 딸기뚱카롱, 고추잡채호빵, 체다 슈크림붕어스낵 ]

```python
import random

# TreeNode 클래스 선언
# ----------------------------------------------------------------------------------------
class TreeNode:
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None
# ----------------------------------------------------------------------------------------

# 전역 변수 선언
# ----------------------------------------------------------------------------------------
memory = []
root = None
dataAry = [ '콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩', 
           '척오리지널 메가 사워', '요아정 요거트컵', '페퍼로니피자주먹밥', 
           '널담 슈톨렌', '딸기마시멜로케이크', '버니공쥬 딸기뚱카롱', 
           '고추잡채호빵', '체다 슈크림붕어스낵']

# ----------------------------------------------------------------------------------------

# 새로 추가된 input부분
# ----------------------------------------------------------------------------------------
print('끝내고 결과를 확인하기 -> e입력')

sellAry = []

while True:
    i = input('판매된 물건을 입력하세요: ') # 입력
    print()
    
    if i.lower() == 'e': # e를 소문자로 바꿔주면서 입력시 브레이크
        break
    
    if i in dataAry: # 입력값이 dataAry에 있으면 저장
        sellAry.append(i)
    else:
        print("입력된 물건이 목록에 없습니다. 다시 입력해 주세요.")

print('오늘 판매된 물건(중복O) --> \n', sellAry)
# ----------------------------------------------------------------------------------------

# 메인코드
# ----------------------------------------------------------------------------------------
# 트리의 첫 번째 노드생성, sellAry의 첫 번째 값을 할당
node = TreeNode()
node.data = sellAry[0]
root = node  # 첫 번째 노드는 트리의 루트
memory.append(node)  # 루트 노드를 메모리 리스트에 추가

# 중복판매된 물품으로 이진탐색트리 생성
for name in sellAry[1:]:
    node = TreeNode() # 새노드 생성
    node.data = name # 노드데이터에 이름쓰기

    current = root
    while True:
        if name == current.data:      # 이미 가지고 있는거 제거? 그냥 넘어가기
            break

        if name < current.data:       # 새노드가 원래 노드보다 작으면
            if current.left == None:  # 왼쪽이 비어있는지 확인
                current.left = node   # 비어있으면 그자리에 들어감
                memory.append(node)   # 만약 처음으로 입력된값이 있으면 메모리에 추가
                break 
            current = current.left    # 왼쪽으로 이동

        else:
            if current.right == None: # 오른쪽이 비어있는지 확인
                current.right = node  # 비어있으면 그자리에 들어감
                memory.append(node)   # 만약 처음으로 입력된값이 있으면 메모리에 추가
                break
            
            current = current.right   # 오른쪽으로 이동
# ----------------------------------------------------------------------------------------

# 중복 없이 출력하는 함수
# ----------------------------------------------------------------------------------------
def preorder(node): # 이진탐색트리 전위순회
    if node == None:
        return
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

print('오늘 판매된 종류(중복X) --> ', end=' ')
preorder(root)
# ----------------------------------------------------------------------------------------
```

4. P473 Self Study 13-2 의 문제. 이진 검색 알고리즘을 풀면서 검색 횟수도 같이 출력하도록 구현하세요.
- 랜덤하게 0과 100000 사이의 숫자 100000개를 생성한 후 0과 100000사이의 숫자 하나를 랜덤하게 뽑아서 찾도록 코드를 작성.

```python
import random

## 이진검색 함수정의
# ----------------------------------------------------------------------------------------
def binSearch(ary, data):
    count = 0
    start = 0
    end = len(ary) - 1 # 리스트의 마지막 인덱스

    while (start <= end): # 시작이 끝보다 커지면 검색실패
        count += 1 # 하나씩 증가
        mid = (start + end) // 2 # 중간값(시작과 끝자리의 반)

        if data == ary[mid]: # 만약 데이터가 미드면
            return count 
        elif data > ary[mid]:
            start = mid + 1 # 뒤쪽에 데이터가 있다
        else:
            end = mid - 1 # 앞쪽에 데이터가 있다

    return count
# ----------------------------------------------------------------------------------------

# 전역변수
# ----------------------------------------------------------------------------------------
ary = random.sample(range(100001), 100000)
ary.sort() # 배열을 오름차순으로 정렬
Random = random.randint(0, 100000) # 0과 100000 사이의 숫자 중 하나를 랜덤하게 선택
count = binSearch(ary, Random) 

# 출력
print(f'## {Random}을(를) 찾았습니다')
print(f'## {count}회 검색함')
# ----------------------------------------------------------------------------------------
```
